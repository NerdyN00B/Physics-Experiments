```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline

# Use pandas to read in the datafile.
# The r before ’filepath’ prevents errors from white spaces in filename.
# This creates a dataframe that I call ’df’

df = pd.read_csv(r'signal.txt', header=None,
              names=['y'])

# The df.head() and df.tail() function can be very helpful
# to figure out what is in the dataframe
print(df.head())

# There is only one column of values named ’y’.
#  To plot this
# raw signal (values vs. index) one could use
signal = df.y
time = np.linspace(0, 1, 1000)

plt.plot(time, signal)
```
